## Simple Movie Recommender System based on user ratings similarity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#Reading in data set
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
u_data = pd.read_csv('data/u.data', sep='\t', names=column_names)
```

```python
u_data.head()
```

```python
movie_titles = pd.read_csv("data/Movie_Id_Titles")
movie_titles.head()
```

```python
full_u_data = u_data.merge(movie_titles, on='item_id')
```

```python
full_u_data.head() # in moment we have a complete data to start analysis
```

### Initial data explenatory analysis...

```python
# Fist step is to group ratings by movie id/title

avg_ratings = full_u_data[['item_id', 'rating']].groupby('item_id').mean()
avg_ratings.head()
```

```python
avg_ratings = avg_ratings.merge(movie_titles, on='item_id')
avg_ratings.head()
```

```python
# Next, count number of raitngs per movie
nb_ratings = full_u_data[['item_id', 'rating']].groupby('item_id').count()
nb_ratings.head()
# full_u_data
```

```python
avg_nb_ratings = avg_ratings.merge(nb_ratings, on='item_id')
```
