# Pandas - unpivot using melt()

Unpivot - means the table is converted from a wide format to a long format.  
Format:  
__df_unpivot = pd.melt(df, id_vars='col1', value_vars=['col2', 'col3', ...])__  

Where:  
__id_vars__: The columns to use as identifiers  
__value_vars__: The columns to unpivot  


```python
import pandas as pd
```

```python
df = pd.DataFrame({'Student' : ['Jenny', 'Singh', 'Charles', 'Richard', 'Veena'],
                   'Date_1': [2000, 7000, 5000, 1500, 2000],
                   'Date_2': [15000, 17000, 27000, 29000, 12000]})
df
```

```python
un_pivot1 = pd.melt(df, id_vars = 'Student', value_vars = ['Date_1', 'Date_2']) 
un_pivot1
```

```python
# giving the new columns better names
un_pivot2 = pd.melt(df, id_vars = 'Student', var_name='Date', value_name='Amount') 
un_pivot2
```

#### Another example

```python
df = pd.DataFrame({'team': ['A', 'B', 'C', 'D', 'E'],
                   'points': [18, 22, 19, 14, 14],
                   'assists': [5, 7, 7, 9, 12],
                   'rebounds': [11, 8, 10, 6, 6]})
df
```

```python
value_cols = df.columns[2:]
value_cols
```

```python
un_pivot3 = pd.melt(df, id_vars = ['team', 'points'], value_vars = value_cols) 
un_pivot3
```
