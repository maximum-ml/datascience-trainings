### Normalization & visualisation

```python
data = {
    'x' : np.concatenate([3 * np.random.randn(500) + 195, 3 * np.random.randn(500) + 205]),
    'y' : np.concatenate([0.1 * np.random.randn(500) + 0.9, 0.1 * np.random.randn(500) + 1.1])
}
data_df = pd.DataFrame(data)
```

```python
data_df.shape
```

```python
data_df.head()
```

```python
# intentionally defined a big ylim to make a ulseless illustration of the distribution
data_df.plot.scatter(x='x', y='y', figsize=(8,8), xlim=(180,220), ylim=(-50, 50))
```

```python
data_df.plot.scatter(x='x', y='y', figsize=(8,8))
```

```python
data_df.mean()
```

```python
data_df.std()
```

```python
data_normalized_df = (data_df - data_df.mean()) / data_df.std()
```

```python
data_normalized_df.head()
```

```python
data_normalized_df.plot.scatter(x='x', y='y', figsize=(8,8))
```
