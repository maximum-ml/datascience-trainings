## 2.  Distributions - displot(), histplot(), kdeplot(), ecdfplot(), rugplot() | previous distplot() DEPRECATED

### 2.1 displot()

```python
sns.displot(car_crashes['alcohol'], kind='hist')
```

```python
sns.displot(penguins, x='body_mass_g', kind='kde', hue='sex')
```

```python
sns.displot(penguins, x='body_mass_g', kind='kde', hue='sex', col='species')
```

```python
sns.displot(tips, x='tip', kind='kde', hue='sex')
```

```python
sns.displot(car_crashes['alcohol'], kind='ecdf') # Empirical Cumulative Distribution
```

### 2.2 histplot()

```python
sns.histplot(car_crashes['alcohol'], bins=20)
```

```python
# A histogram with KDE (kernel density estimation) line presented by using kde flag set to True.
plt.figure(figsize = (12,6))
sns.histplot(car_crashes['alcohol'], bins=50, kde=True)
```

```python
sns.rugplot(data=penguins, x='body_mass_g', y='bill_length_mm', hue='sex', height=0.1) # shows only bars on the axis (dywan, koc)
```

## 3. Relational Plots  - scatterplot(), lineplot()

### 3.1. scatterplot()

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

### 3.2. lineplot()

```python
x = np.arange(-9.99, 9.99, 0.01)
y = 2 * np.square(x) + x + 1 
```

```python
sns.lineplot(x=x, y=y)
```

### 4.2. countplot()

```python
sns.countplot(data=penguins, x='sex')
```

### 4.3. barplot()

```python
sns.barplot(data=penguins, x='species', y='body_mass_g', hue='sex', estimator=np.mean) 
# presents the mean value for mass with variance (black strips)
```

### 4.4. boxplot()

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g', hue='sex') # represents the mean, std deviation, max/min data & outliers
```

### 4.5. stripplot()

```python
plt.figure(figsize=(8,6))
sns.stripplot(data=penguins, x='species', y='body_mass_g', hue='sex')
```

```python
plt.figure(figsize=(8,6))
sns.stripplot(data=penguins, x='species', y='body_mass_g', hue='sex', dodge=True)
```

### 4.6. swarmplot()

```python
sns.swarmplot(data=penguins, x='sex', y='body_mass_g')
```
